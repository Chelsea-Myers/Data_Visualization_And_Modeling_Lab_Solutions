```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv('Employee_productivity.csv')

df.head()
```

## Relationships with productivity

```python
### Relationship of hours worked to productivity
```

```python
sns.regplot(x='HoursWorked', y='ProductivityScore', data=df)
plt.title('Relationship of Hours Worked to Productivity')
plt.show()
```

```python
X = sm.add_constant(df['HoursWorked'])
y = df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

### Relationship of years at company to productivity

```python
sns.regplot(x='YearsAtCompany', y='ProductivityScore', data=df)
plt.title('Relationship of Years at Company to Productivity')
plt.show()
```

```python
X = sm.add_constant(df['YearsAtCompany'])
y = df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

### Relationship of focus to productivity

```python
sns.regplot(x='FocusScore', y='ProductivityScore', data=df)
plt.title('Relationship of Focus to Productivity')
plt.show()
```

```python
X = sm.add_constant(df['FocusScore'])
y = df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

Focus is the most predictive of productivity.  Years at the company and hours worked don't seem to matter in terms of worker productivity.

## Adding in other factors

### Meetings per week

```python
#How many meetings per week do people have?

df['TeamMeetingsPerWeek'].value_counts()
```

```python
#Combine 6 or more meetings in a single category
#Note, students probably won't do this but we do talk about combining small categories together in class


df.loc[ df['TeamMeetingsPerWeek'] > 6, 'TeamMeetingsPerWeek'] = 6

df['TeamMeetingsPerWeek'].value_counts()
```

```python
sns.scatterplot(x='FocusScore', y='ProductivityScore', hue='TeamMeetingsPerWeek', data=df)
plt.title('Relationship of Focus and Number of Meetings to Productivity')
plt.show()
```

```python
X = sm.add_constant(df[['FocusScore', 'TeamMeetingsPerWeek']])
y = df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

### Desk location

```python
#Where do people work?

df['DeskLocation'].value_counts()
```

```python
#Simple recode of desk location
#Students can do this in any reasonable way.  We have not talked about one-hot encoding at this point.

df.loc[df['DeskLocation'] == 'Remote', 'DeskLocation_num' ] = 0
df.loc[df['DeskLocation'] == 'Private Office', 'DeskLocation_num' ] = 1
df.loc[df['DeskLocation'] == 'Open Plan', 'DeskLocation_num' ] = 2

df['DeskLocation_num'].value_counts()
```

```python
sns.scatterplot(x='FocusScore', y='ProductivityScore', hue='DeskLocation_num', data=df)
plt.title('Relationship of Focus and Work Arrangement to Productivity')
plt.legend(['Open Plan', 'Private Office', 'Remote'], title='Work Arrangement')
plt.show()
```

```python
X = sm.add_constant(df[['FocusScore', 'DeskLocation_num']])
y = df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

### Department

```python
#Where do people work?

df['Department'].value_counts()
```

```python
#Simple recode of department
#Students can do this in any reasonable way.  We have not talked about one-hot encoding at this point.

df.loc[df['Department'] == 'HR', 'Department_num' ] = 0
df.loc[df['Department'] == 'Engineering', 'Department_num' ] = 1
df.loc[df['Department'] == 'Sales', 'Department_num' ] = 2
df.loc[df['Department'] == 'Marketing', 'Department_num' ] = 3

df['Department_num'].value_counts()
```

```python
sns.scatterplot(x='FocusScore', y='ProductivityScore', hue='Department_num', data=df)
plt.title('Relationship of Focus and Department to Productivity')
plt.legend(['Marketing', 'Sales', 'Engineering', 'HR'], title='Department')
plt.show()
```

```python
X = sm.add_constant(df[['FocusScore', 'Department_num']])
y = df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

## Conclusion

Focus is the most important factor contributing to productivity.  Differences in focus explain about 67.8% of the differences in employee productivity.  After taking focus into account, desk location, number of meetings and department don't really matter.
