```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic_passengers.csv')

df.head()
```

## Categorizing age and fare

I'm going to categorize age as <= 12 (children), 13 - 25 (youth), 25 - 45 (adults), 45 + (older adults)

I'm going to categorize fare based in quartiles.

```python
#Age in categories
#Note that students can chose any reasonable categories and use any reasonable coding approach to achieve this

df.loc[df['Age'] <= 12 , 'Age_cat' ] = 'Child'
df.loc[(df['Age'] > 12) & (df['Age'] <= 25) , 'Age_cat' ] = 'Youth'
df.loc[(df['Age'] > 25) & (df['Age'] <= 45) , 'Age_cat' ] = 'Adult'
df.loc[(df['Age'] > 45), 'Age_cat' ] = 'Older Adult'

df['Age_cat'].value_counts()
```

```python
#Fare in quartiles
#Note that students can chose any reasonable categories and use any reasonable coding approach to achieve this

df['Fare_cat'] = pd.qcut(df['Fare'], q=4, labels=['Low', 'Low-Medium', 'High-Medium', 'High'])

df['Fare_cat'].value_counts()
```

## Associations with Survival

### Age

```python
pd.crosstab(df['Age_cat'], df['Survived']).plot(kind='bar')

plt.title('Relationship of Age to Survival on the Titanic')
plt.legend(['Perished', 'Survived'], title='Outcome')
plt.ylabel('Number of Passengers')
plt.xlabel('Age')
plt.xticks(rotation=0)

plt.show()
```

Children were the only group who were more likely to survive than not.  Youth and adults were least likely to survive.

### Fare

```python
pd.crosstab(df['Fare_cat'], df['Survived']).plot(kind='bar')

plt.title('Relationship of Fare to Survival on the Titanic')
plt.legend(['Perished', 'Survived'], title='Outcome')
plt.ylabel('Number of Passengers')
plt.xlabel('Fare')
plt.xticks(rotation=0)

plt.show()
```

People who paid a low fare were the least likely to survive and people who paid a high fare were most likely to surivive.

## Sex

```python
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar')

plt.title('Relationship of Passenger Sex to Survival on the Titanic')
plt.legend(['Perished', 'Survived'], title='Outcome')
plt.ylabel('Number of Passengers')
plt.xlabel('Passenger Class')
plt.xticks([0,1], ['Female', 'Male'], rotation=0)

plt.show()
```

Female passengers are much more likely to survive than male passengers

### Passenger Class

```python
pd.crosstab(df['Pclass'], df['Survived']).plot(kind='bar')

plt.title('Relationship of Passenger Class to Survival on the Titanic')
plt.legend(['Perished', 'Survived'], title='Outcome')
plt.ylabel('Number of Passengers')
plt.xlabel('Passenger Class')
plt.xticks([0,1,2], ['1st', '2nd', '3rd'], rotation=0)

plt.show()
```

People in first class were more likely to survive but people in third class were much less likely to surivive.

### Port of Departure

```python
pd.crosstab(df['Embarked'], df['Survived']).plot(kind='bar')

plt.title('Relationship of Port of Departure to Survival on the Titanic')
plt.legend(['Perished', 'Survived'], title='Outcome')
plt.ylabel('Number of Passengers')
plt.xlabel('Port of Departure')
plt.xticks([0,1,2], ['Cherbourg', 'Queenstown', 'Southampton'], rotation=0)

plt.show()
```

People in Cherbourg were most likely to survive and people from Southampton were least likely to survive.

### Travel Companions

I'm going to categorize travel companions into people with no travel companions vs people with at least one travel companion (either parents/children or siblings/spouse).

Note: Students can choose to do this in any reasonable way.

```python
df['Companions'] = df['Parch'] + df['SibSp']

df.loc[df['Companions'] == 0 , 'Companions_cat' ] = 'Alone'
df.loc[df['Companions'] > 0 , 'Companions_cat' ] = 'With Companions'
```

```python
pd.crosstab(df['Companions_cat'], df['Survived']).plot(kind='bar')

plt.title('Relationship of Travel Companions to Survival on the Titanic')
plt.legend(['Perished', 'Survived'], title='Outcome')
plt.ylabel('Number of Passengers')
plt.xlabel('')
plt.xticks(rotation=0)

plt.show()
```

People traveling alone were much less likely to survive than people traveling with companions.
