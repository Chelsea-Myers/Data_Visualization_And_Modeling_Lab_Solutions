```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

df.head()
                
```

## Dataset

For this lab, I'm using the sleep health and lifestyle dataset. The Sleep Health and Lifestyle Dataset contains 374 rows and 13 columns, capturing details about sleep patterns, lifestyle habits, and related health indicators. 

My target feature is sleep disorder, which is categorized as none, sleep apnea or insomnia.  For the purposes of this lab, I'm going to combine sleep apnea and insomnia into a single category where sleep disorder = yes (1) and if neither of those are true then sleep disorder = no (0).

The 5 features of interest that I'm going to look at are gender, age, physical activity level, occupation, and quality of sleep.

**Note: Students can choose any dataset and figures as long as they follow the instructions above.**

```python
## Fill missing data with 'none' for sleep disorder'

df['Sleep Disorder'] = df['Sleep Disorder'].fillna("None")

df["Sleep Disorder"].value_counts()
```

```python
# Create new categorical feature where sleep disorder = 0 (no sleep disorder) or 1 (sleep disorder - insomnia/sleep apnea)
df['Sleep_Disorder_cat'] = '1'
df.loc[df['Sleep Disorder'] == "None", "Sleep_Disorder_cat"] ='0'

df['Sleep_Disorder_cat'].value_counts()
```

## Sleep Disorders and Gender

```python
pd.crosstab(df['Gender'], df['Sleep_Disorder_cat']).plot(kind='bar')
plt.title('Relationship of Gender to the Presence of a Sleep Disorder')
plt.ylabel('Number of Participants')
plt.xticks(rotation=0)
plt.legend(['No', 'Yes'], title='Presence of Sleep Disorder')

plt.show()
```

In this dataset, female participants were more likely to have a sleep disorder than male participants.

## Sleep Disorders and Age

```python
sns.boxplot(x='Age', y='Sleep_Disorder_cat', data=df)

plt.title('Relationship of Age to the Presence of a Sleep Disorder')
plt.ylabel('Presence of Sleep Disorder')
plt.yticks([0,1], ['No', 'Yes'])

plt.show()
```

People with sleep disorders tend to be older than people without sleep disorders.  The median age of those with sleep disorders is about 45 and the median age of people without sleep disorders is about 37.

## Sleep Disorders and Physical Activity Level

```python
sns.boxplot(x='Physical Activity Level', y='Sleep_Disorder_cat', data=df)

plt.title('Relationship of Physical Activity Level to the Presence of a Sleep Disorder')
plt.ylabel('Presence of Sleep Disorder')
plt.yticks([0,1], ['No', 'Yes'])

plt.show()
```

```python
df.groupby('Sleep_Disorder_cat')['Physical Activity Level'].describe()
```

The relationship of physical activity to the presence of a sleep disorder is harder to interpret.  The mean physical activity in both groups is very similar.  However, the median physical activity is lower in those with a sleep disorder than in those without.  It's not conclusive that physical activity is related to having a sleep disorder.

## Sleep Disorders and Occupation

```python
pd.crosstab(df['Occupation'], df['Sleep_Disorder_cat']).plot(kind='bar')
plt.title('Relationship of Occupation to the Presence of a Sleep Disorder')
plt.ylabel('Number of Participants')
plt.xticks(rotation=90)
plt.legend(['No', 'Yes'], title='Presence of Sleep Disorder')

plt.show()
```

Nurses, salespeople, sales representatives, and teachers were more likely to have a sleep disorder than those in other professions.

## Sleep Disorders and Quality of Sleep

```python
sns.boxplot(x='Quality of Sleep', y='Sleep_Disorder_cat', data=df)

plt.title('Relationship of Quality of Sleep to the Presence of a Sleep Disorder')
plt.ylabel('Presence of Sleep Disorder')
plt.yticks([0,1], ['No', 'Yes'])

plt.show()
```

```python
df.groupby('Sleep_Disorder_cat')['Quality of Sleep'].describe()
```

The quality of sleep tended to be much higher in those without sleep disorders than in those with sleep disorders.

## Conclusion

In this analysis, I have found that age, gender and occupation were related to an individual having a sleep disorder while physical activity level was not.  Individuals who had a sleep disorder tended to have a lower quality of sleep than those who did not have a sleep disorder.

Women should be on the lookout for sleep disorders as they age, particularly if they are teachers, nurses or work in sales.
