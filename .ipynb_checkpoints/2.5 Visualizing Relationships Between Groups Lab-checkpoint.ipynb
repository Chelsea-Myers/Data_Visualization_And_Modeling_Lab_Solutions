# Lab Instructions

You are part of a data team for a financial institution that uses an AI model to determine who will and will not be granted a loan.  Recently, there have been complaints from customers that the model unfairly discriminates based on the applicant's age and sex. A court has ordered that the financial institution must provide information about how the model works.  

Data from recent loan applications is given in `loan_data.csv`.  The data dictionary is below:

| Feature              | Description | Data Type |
| :---------------- | :------: | ----: |
|person_age|Age of the person|Float
|person_gender|Gender of the person|Categorical
|person_education|Highest education level|Categorical
|person_income|Annual income|Float
|person_emp_exp|Years of employment experience|Integer
|person_home_ownership|Home ownership status (e.g., rent, own, mortgage)|Categorical
|loan_amnt|Loan amount requested|Float
|loan_intent|Purpose of the loan|Categorical
|loan_int_rate|Loan interest rate|Float
|loan_percent_income|Loan amount as a percentage of annual income|Float
|cb_person_cred_hist_length|Length of credit history in years|Float
|credit_score|Credit score of the person|Integer
|previous_loan_defaults_on_file|Indicator of previous loan defaults|Categorical
|loan_status (target variable)|Loan approval status: 1 = approved; 0 = rejected|Integer

Visualize the relationship between age, sex and 5 other features with `loan_status`.  Which features seem to drive the model's decision to grant a loan?  Which seem unrelated?  Should the financial institution be concerned about potential legal trouble?  Make sure to justify your answers with specific references to your visualizations.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv('loan_data.csv')

df.head()
```

## Relationship of age to loan status

```python

df['loan_status'] = df['loan_status'].astype('category')

sns.boxplot(x='person_age', y='loan_status', data=df)
plt.title('Relationship of Age to Loan Status')
plt.ylabel('Loan Status')
plt.yticks([0,1], ['Denied', 'Approved'])
plt.xlabel('Age')
plt.show()
```

No one over age 60 was approved for a loan - this seems suspicious.

## Relationship of gender to loan status

```python
crosstab = pd.crosstab(df['person_gender'], df['loan_status'])

crosstab.plot(kind='bar')
plt.title('Relationship of Gender to Loan Status')
plt.legend(['Denied', 'Accepted'], title='Loan Status')
plt.ylabel('Number of Applicants')
plt.xticks(rotation=0)
plt.xlabel('')
plt.show()
```

About the same proportion of male and female applicants had their loans accepted.

## Relationship of education to loan status

Note that students can select any 5 other features.  These are just examples.

```python
crosstab = pd.crosstab(df['person_education'], df['loan_status'])

crosstab.plot(kind='bar')
plt.title('Relationship of Education Level to Loan Status')
plt.legend(['Denied', 'Accepted'], title='Loan Status')
plt.ylabel('Number of Applicants')
plt.xticks(rotation=0)
plt.xlabel('Education Level')
plt.show()
```

About the same proportion of applicants from each educational category had their loans accepted.

## Relationship of income to loan status

Note that students can select any 5 other features.  These are just examples.

```python
sns.boxplot(x='person_income', y='loan_status', data=df)
plt.title('Relationship of Income to Loan Status')
plt.ylabel('Loan Status')
plt.yticks([0,1], ['Denied', 'Approved'])
plt.xlabel('Income')
plt.show()
```

People with higher income were more likely to be accepted.  This seems normal.

## Relationship of credit score to loan status

Note that students can select any 5 other features.  These are just examples.

```python
sns.boxplot(x='credit_score', y='loan_status', data=df)
plt.title('Relationship of Credit Score to Loan Status')
plt.ylabel('Loan Status')
plt.yticks([0,1], ['Denied', 'Approved'])
plt.xlabel('Income')
plt.show()
```

People with a higher credit score were more likely to have their loan accepted, which seems normal.

## Relationship of previous pefaults to loan status

```python
crosstab = pd.crosstab(df['previous_loan_defaults_on_file'], df['loan_status'])

crosstab.plot(kind='bar')
plt.title('Relationship of Previous Defaults to Loan Status')
plt.legend(['Denied', 'Accepted'], title='Loan Status')
plt.ylabel('Number of Applicants')
plt.xticks(rotation=0)
plt.xlabel('Previous Defaults')
plt.show()

```

No one with a previous default was granted a loan.  That seems reasonable.

## Relationship of intended use to loan status

```python
crosstab = pd.crosstab(df['loan_intent'], df['loan_status'])

crosstab.plot(kind='bar')
plt.title('Relationship of Loan Use to Loan Status')
plt.legend(['Denied', 'Accepted'], title='Loan Status')
plt.ylabel('Number of Applicants')
plt.xlabel('Loan Use')
plt.show()

```

People are granted loans at different rates for differnet uses; however, this isn't a protected class so it's at the bank's discretion to do so.

## Important Features

It looks like an applicant's gender and education level are not related to whether or not they are granted alone.  However, age, income, credit score, use and previous defaults are.  It's totally reasonable for a bank to use things that are directly related to creditworthiness to determine if someone gets a loan; however, the fact that they didn't grant a loan to anyone over age 60 might be cause for concern.
