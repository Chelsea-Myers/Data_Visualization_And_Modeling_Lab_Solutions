```python
import pandas as pd

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

df.head()
```

## Beats per Minute

```python
df['bpm'].plot(kind='hist',
               title='Distribution of Beats Per Minute in Songs on Spotify',
               xlabel='Beats per Minute')
```

```python
#Note that a box plot would also be appropriate here although students are typically not familiar with that kind
#of visualization yet

df['bpm'].describe()
```

The mean and median tempo of songs in the Spotify dataset is about 120 bpm with a range of 65 to 206 bpm.  Most songs have a tempo between about 100 and 140 bpm.

## Key

```python
#Note a pie chart would also be acceptable here

df['key'].value_counts().plot(kind='bar',
                              title='Distribution of Key in Songs on Spotify')
```

```python
df['key'].value_counts()
```

The most common key in the Spotify dataset is C#.
